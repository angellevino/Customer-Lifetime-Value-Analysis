```python
import pandas as pd
import seaborn as sns
data = pd.read_csv("customer_acquisition_data.csv")
data.head(5)
```

```python
#cost Calculation for each channel
total_cost = data.groupby("channel")["cost"].mean()
sns.barplot(x=total_cost.index, y=total_cost)
```

```python
# Revenue generated for each marketing channel 
revenue_generate = data.groupby("channel")["revenue"].sum()
sns.barplot(x=revenue_generate.index, y=revenue_generate)
```

```python
# ROI Calculation
data["ROI"] = data["revenue"] / data["cost"]
data.head(5)

roi_channel = data.groupby("channel")["ROI"].mean()
sns.barplot(x=roi_channel.index, y= roi_channel)
```

```python
# CR by channel 
CR_channel = data.groupby("channel")["conversion_rate"].mean()
sns.barplot(x=CR_channel.index, y=CR_channel)
```

```python
# Customer lifetime for each channel CLTV = (revenue  cost) * conversion_rate / cost
data["CLTV"] = (data["revenue"]- data["cost"])*data["conversion_rate"]/data["cost"]
CLTV_channel = data.groupby("channel")["CLTV"].mean()
data.head(5)
sns.barplot(x=CLTV_channel.index, y=CLTV_channel)
```
